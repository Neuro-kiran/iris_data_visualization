## Data Visualization of iris data 

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(iris['sepal_length'], iris['sepal_width'])
plt.title("Scatter Plot")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
iris_grouped = iris.groupby('species').mean()
iris_grouped['sepal_length'].plot(kind='bar')
plt.title("Bar Plot")
plt.xlabel("Species")
plt.ylabel("Average Sepal Length (cm)")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
iris_grouped['sepal_length'].plot(kind='barh')
plt.title("Horizontal Bar Plot")
plt.ylabel("Species")
plt.xlabel("Average Sepal Length (cm)")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
species_counts = iris['species'].value_counts()
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(iris['sepal_length'], bins=20, edgecolor='black')
plt.title("Histogram")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='sepal_length', data=iris)
plt.title("Box Plot")
plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(iris['sepal_length'], label='sepal Length')
plt.plot(iris['sepal_width'], label='sepal width')
plt.title("Line ploat")
plt.xlabel("sample Index")
plt.ylabel("Length/width (cm)")
plt.legend()
plt.show()
```
